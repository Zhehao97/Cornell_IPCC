```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt?
```

```python
inpt_path = 'Cornell Data.xlsx'
```

```python
Tbill = pd.read_excel(io=inpt_path, sheet_name='T-Bill')
TbillETF = pd.read_excel(io=inpt_path, sheet_name='1-3 MONTH ETF')

# convert yield to price
Tbill['Mid Price'] = Tbill['Mid Price'].apply(lambda x:100/(1+x/12))
```

```python
AggIndex = pd.read_excel(io=inpt_path, sheet_name='Bloomberg Agg Index')
AggETF = pd.read_excel(io=inpt_path, sheet_name='Barclays Agg ETF')
```

```python
SP500 = pd.read_excel(io=inpt_path, sheet_name='S&P500')
```

```python
Russell1000 = pd.read_excel(io=inpt_path, sheet_name='Russell 1000')
Russell1000ETF = pd.read_excel(io=inpt_path, sheet_name='russell 1000 etf')
```

```python
EmergingETF = pd.read_excel(io=inpt_path, sheet_name='SPDR ETF')
```

### 1. Portfolio using Index as Instruments

```python
Portfolio_Classic = Tbill.merge(AggIndex, on='Date')
Portfolio_Classic = Portfolio_Classic.merge(SP500, on='Date')
Portfolio_Classic = Portfolio_Classic.merge(Russell1000, on='Date')
```

```python
Portfolio_Classic.columns = ['Date', 'T-Bill', 'US AggIndex', 'S&P 500', 'Russell 1000']
```

```python
Portfolio_Classic = Portfolio_Classic.set_index('Date')
Portfolio_Classic = Portfolio_Classic.sort_index(ascending=True)
```

```python
# calculate the daily return for each Index instrument 
PotfReturn_Classic = Portfolio_Classic.apply(lambda x:(x/x.shift()-1))
```

```python
PotfReturn_Classic.head()
```

```python
def portfolioReturns(Portfolio, PotfReturn, initi, equityp, bondp):
    value = initi # initial assets
    lv = [initi]
    years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
    r = Portfolio.iloc[0, 3] # Russell 1000
    agg = Portfolio.iloc[0, 1] # US AggIndex
    t = Portfolio.iloc[0, 0] # T-Bill
    perc_r = equityp
    perc_agg = bondp*0.8
    perc_t = bondp*0.2
    value_r = initi*perc_r 
    value_agg = initi*perc_agg
    value_t = initi*perc_t
    shares_r = value_r/r
    shares_agg = value_agg/agg
    shares_t = value_t/t
    for year in years: 
        for index, row in PotfReturn[PotfReturn.index.year == year].iterrows():
            r = Portfolio.loc[index, 'Russell 1000']
            agg = Portfolio.loc[index, 'US AggIndex']
            t = Portfolio.loc[index, 'T-Bill']
            
            value_r = shares_r*r
            value_agg = shares_agg*agg
            value_t = shares_t*t
            value = value_r + value_agg + value_t
            lv.append(value)
        
        value_r = value*perc_r
        value_agg = value*perc_agg
        value_t = value*perc_t
        shares_r = value_r/r
        shares_agg = value_agg/agg
        shares_t = value_t/t
        
    for index, row in PotfReturn[PotfReturn.index.year == 2020].iterrows():
        r = Portfolio.loc[index, 'Russell 1000']
        agg = Portfolio.loc[index, 'US AggIndex']
        t = Portfolio.loc[index, 'T-Bill']
        
        value_r = shares_r*r
        value_agg = shares_agg*agg
        value_t = shares_t*t
        value = value_r + value_agg + value_t
        lv.append(value)
        
    return lv    
    
```

```python
for i in range(1,10):
    equity_weight = 0.1 * i
    bonds_weight = 0.1 * (10 - i)
    ll = portfolioReturns(Portfolio_Classic, PotfReturn_Classic.dropna(), 1000, equity_weight, bonds_weight)
    col_name = (i/(10-i))
    PotfReturn_Classic[col_name] = ll
    PotfReturn_Classic[col_name] = PotfReturn_Classic[col_name]/PotfReturn_Classic[col_name].shift()-1
```

```python
PotfReturn_Classic
```

```python
# calculate the rolling std of daily return for each Index instrument 
PotfStd_Classic = PotfReturn_Classic.std()
```

```python
PotfStd_Classic_Annual = PotfStd_Classic * np.sqrt(250)
PotfStd_Classic_Annual = PotfStd_Classic_Annual[4:]
```

```python
PotfStd_Classic_Annual
```

```python
x = np.array(PotfStd_Classic_Annual.values)
y = np.array(PotfStd_Classic_Annual.index)

plt.plot(x, y, drawstyle='steps')
plt.plot(x, y)
plt.xlabel('Std of Different Ratio')
plt.ylabel('Ratio (Equity/Bond)')
plt.title('Index')
plt.show()
```

### 2. Portfolio using ETFs as Instruments

```python
Portfolio_ETF = TbillETF.merge(AggETF, on='Date', how='outer')
Portfolio_ETF = Portfolio_ETF.merge(EmergingETF, on='Date')
Portfolio_ETF = Portfolio_ETF.merge(Russell1000ETF, on='Date')
```

```python
Portfolio_ETF
```

```python
Portfolio_ETF.columns = ['Date', 'T-Bill ETF', 'US AggETF', 'Emerging Market ETF', 'Russell 1000 ETF']
```

```python
Portfolio_ETF = Portfolio_ETF.set_index('Date')
Portfolio_ETF = Portfolio_ETF.sort_index(ascending=True)
```

```python
# calculate the daily return for each ETF instrument 
PotfReturn_ETF = Portfolio_ETF.apply(lambda x:(x/x.shift()-1))
```

```python
def portfolioReturns_v2(Portfolio, PotfReturn, initi, equityp, bondp):
    value = initi # initial assets
    lv = [initi]
    years = [2018, 2019]
    r = Portfolio.iloc[0, 3] # Russell 1000
    agg = Portfolio.iloc[0, 1] # US AggIndex
    t = Portfolio.iloc[0, 0] # T-Bill
    perc_r = equityp
    perc_agg = bondp*0.8
    perc_t = bondp*0.2
    value_r = initi*perc_r 
    value_agg = initi*perc_agg
    value_t = initi*perc_t
    shares_r = value_r/r
    shares_agg = value_agg/agg
    shares_t = value_t/t
    for year in years: 
        for index, row in PotfReturn[PotfReturn.index.year == year].iterrows():
            r = Portfolio.loc[index, 'Russell 1000 ETF']
            agg = Portfolio.loc[index, 'US AggETF']
            t = Portfolio.loc[index, 'T-Bill ETF']
            
            value_r = shares_r*r
            value_agg = shares_agg*agg
            value_t = shares_t*t
            value = value_r + value_agg + value_t
            lv.append(value)
        
        value_r = value*perc_r
        value_agg = value*perc_agg
        value_t = value*perc_t
        shares_r = value_r/r
        shares_agg = value_agg/agg
        shares_t = value_t/t
        
    for index, row in PotfReturn[PotfReturn.index.year == 2020].iterrows():
        r = Portfolio.loc[index, 'Russell 1000 ETF']
        agg = Portfolio.loc[index, 'US AggETF']
        t = Portfolio.loc[index, 'T-Bill ETF']
        
        value_r = shares_r*r
        value_agg = shares_agg*agg
        value_t = shares_t*t
        value = value_r + value_agg + value_t
        lv.append(value)
        
    return lv    
    
```

```python
PotfReturn_ETF_2 = pd.DataFrame()
for i in range(1,10):
    equity_weight = 0.1 * i
    bonds_weight = 0.1 * (10 - i)
    ll = portfolioReturns_v2(Portfolio_ETF.dropna(how='any'), PotfReturn_ETF.dropna(how='any'), 1000, equity_weight, bonds_weight)
    col_name = (i/(10-i))
    PotfReturn_ETF_2[col_name] = ll
    PotfReturn_ETF_2[col_name] = PotfReturn_ETF_2[col_name]/PotfReturn_ETF_2[col_name].shift()-1
```

```python
PotfReturn_ETF_2
```

```python
# calculate the rolling std of daily return for each ETF instrument 
PotfStd_ETF = PotfReturn_ETF_2.std()
```

```python
PotfStd_ETF_Annual = PotfStd_ETF * np.sqrt(250)
PotfStd_ETF_Annual
```

```python
xx = np.array(PotfStd_ETF_Annual.values)
yy = np.array(PotfStd_ETF_Annual.index)
```

```python
plt.figure(figsize=(16,8))
plt.plot(x, y, drawstyle='steps', label='Index steps')
plt.plot(x, y, label='Index curve')

plt.plot(xx, yy, drawstyle='steps', label='ETF steps')
plt.plot(xx, yy, label='ETF curve')

plt.legend()
plt.xlabel('Std of Different Ratio')
plt.ylabel('Ratio (Equity/Bond)')
plt.title('Index and ETF')
plt.show()
```
